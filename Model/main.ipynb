```python
from keras.utils import to_categorical
```

```python
y_train_Digonasis = to_categorical(y_train['Diagnosis'].factorize()[0],num_classes=4)
y_train_stage = to_categorical(y_train['Disease Stage'].factorize()[0],num_classes=5)
y_train = [y_train_Digonasis,y_train_stage]
```

```python
y_train
```

```python
from keras.layers import Dense,Input
from keras.models import Model
```

```python
data['Disease Stage'].unique()
```

```python
inputs = Input(shape=(27,))
x = Dense(64,activation='relu') (inputs)
x = Dense(32,activation='relu') (x)
Digonasis_output = Dense(4,activation='softmax',name='Digonasis_output')(x)
stage_output = Dense(5,activation='softmax',name='stage_output')(x)
```

```python
model = Model(inputs=inputs,outputs=[Digonasis_output,stage_output] )
```

```python
model.compile(optimizer='adam',loss={'Digonasis_output':"categorical_crossentropy",'stage_output':"categorical_crossentropy"},metrics=['accuracy','accuracy'])
```

```python
model.fit(x_train,{'Digonasis_output':y_train_Digonasis,'stage_output':y_train_stage},epochs=50,batch_size=62)
```
